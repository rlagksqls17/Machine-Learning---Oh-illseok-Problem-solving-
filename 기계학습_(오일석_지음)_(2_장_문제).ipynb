1. 행렬 A
> 1 -2 3 5  
> 2 2 -1 0  
> 3 0 1 2  
> 1 0 2 0  
의 2A, A^t, A^-1을 쓰시오, 또한 A의 계수를 쓰시오.

```
import numpy as np
import numpy.linalg as lin
# 행렬 A는 다음과 같이 표기할 수 있음
pro_A = np.array([[1, -2, 3, 5], [2, 2, -1, 0], [3, 0, 1, 2], [1, 0, 2, 0]])
print(f"행렬 A : \n {pro_A}")
# 2A를 구하려면 다음과 같이 하면 됨
# np.full() : 배열에 사용자가 지정한 값을 넣음
pro_sample = np.full((4, 4), 2)
print(f"행렬 2A : \n {pro_A + pro_sample}") # 2A

# A의 전치행렬을 구하려면 다음과 같이 함
# 행렬.T : 전치행렬을 구해줌
print(f"A의 전치행렬 : \n {pro_A.T}")

# A의 역행렬을 구하려면 다음과 같이 함
# lin.inv() : 역행렬을 구해줌
print(f"A의 역행렬 : \n {lin.inv(pro_A)}")

# A의 계수(Rank)를 구하려면 다음과 같이 함
# lin.matrix_rank(행렬)
print(f"A의 계수 : {lin.matrix_rank(pro_A)}")
```

2. A = 
> (1 -2)  
> (2  2)  

의 A^1/2과 A^-1/2를 구하시오.

```
'''
p x p 양정치행렬 A에 대해 
A^1/2 * A^1/2 = A 를 만족하는 행렬 A^1/2를 제곱근 행렬이라 한다.
'''
import numpy as np
import numpy.linalg as lin
from scipy import linalg

# scipy의 linalg.sqrtm(행렬)을 사용하여 제곱근 행렬을 계산한다.
pro_A = np.array([[1, -2], [2, 2]])
AOneOverTwo = linalg.sqrtm(pro_A)
print(f"A^1/2 = \n{AOneOverTwo}")
print(f"A^-1/2 = \n{lin.inv(AOneOverTwo)}")
```

5. 놈을 계산하시오.  
(1) x = (3 -4 -1.2 0 2.3)^T의 1차, 2차, 3차 놈과 최대 놈  
(2) [[2, 1], [1, 5], [4, 1]]의 프로베니우스 놈  



```
x = [3, -4, -1.2, 0, 2.3]

def LpNorm(X, p):
    sum = 0
    for a in X:
        if a < 0:
            a = a * (-1)
        sum += (a ** p)
    return sum ** (1/p)

def FrobNorm(X):
    sum = 0
    for a in X:
        sum += (a**2)
    return (sum ** 0.5)

print(f"1차 놈 = {LpNorm(x, 1)}\n2차 놈 = {LpNorm(x,2)}\n3차 놈 = {LpNorm(x, 3)}\n최대 놈 = {-x[1]}")
print(f"프로베니우스 놈 = {FrobNorm(x)}")
```

8. m = 8일 때 p = 0.1, 0.5, 0.8 각각의 이항 분포를 그림으로 그리시오.  

문제를 좀 더 자세하게 분해해보자.  

m = 8이란 뜻은 시행이 8번이라는 것이다. 이때 p = 0.1이라면 시행했을 때 확률이 10% 라는 것이다. 따라서 다음의 값을 분포로 그린다.  

```
import numpy as np
import matplotlib.pyplot as plt
from math import factorial

# Probability density of the binomial distribution
def bin_dist(k, n, p):
    nck = factorial(n) / (factorial(k) * factorial(n - k))
    pd = nck * p**k * (1-p)**(n-k)
    return pd

x = np.arange(9)
pd1 = np.array([bin_dist(k, 8, 0.1) for k in range(9)])
plt.ylim(0, 1)
plt.text(12.5, 0.28, 'n, p = 8, 0.1')
plt.bar(x, pd1, color='lightcoral')
plt.show()


x = np.arange(9)
pd1 = np.array([bin_dist(k, 8, 0.5) for k in range(9)])
plt.ylim(0, 1)
plt.text(12.5, 0.28, 'n, p = 8, 0.5')
plt.bar(x, pd1, color='lightcoral')
plt.show()


x = np.arange(9)
pd1 = np.array([bin_dist(k, 8, 0.8) for k in range(9)])
plt.ylim(0, 1)
plt.text(12.5, 0.28, 'n, p = 15, 0.8')
plt.bar(x, pd1, color='lightcoral')
plt.show()
```

확률 분포도 다양한 종류의 분포가 있는데, 여기서는 그저 a가 일어날 확률만 명시되어 있으므로 베르누이 분포를 활용해보자.

```
# Scipy의 서브 패키지에 있는 bernoulli 클래스를 이용하자
import scipy as sp
from scipy.stats import bernoulli

# p = 0.4
p = 0.4
rv = sp.stats.bernoulli(p)
xx = [0, 1] # a 사건 일어나면 xx = 1
plt.bar(xx, rv.pmf(xx), align = "center")
plt.xlim(-1, 2)
plt.ylim(0, 1)
plt.xticks([0, 1], ["X=0", "X=1"])
plt.ylabel("P(x)")
plt.show()

p = 0.5
rv = sp.stats.bernoulli(p)
xx = [0, 1] # a 사건 일어나면 xx = 1
plt.bar(xx, rv.pmf(xx), align = "center")
plt.xlim(-1, 2)
plt.ylim(0, 1)
plt.xticks([0, 1], ["X=0", "X=1"])
plt.ylabel("P(x)")
plt.show()

p = 0.6
rv = sp.stats.bernoulli(p)
xx = [0, 1] # a 사건 일어나면 xx = 1
plt.bar(xx, rv.pmf(xx), align = "center")
plt.xlim(-1, 2)
plt.ylim(0, 1)
plt.xticks([0, 1], ["X=0", "X=1"])
plt.ylabel("P(x)")
plt.show()
```

(2) 각 경우의 엔트로피를 구하시오.  

엔트로피는 확률분포가 가지는 정보의 확신도 혹은 정보량을 수치로 표현한 것이다. 확률분포에서 특정한 값이 나올 확률이 높아지고 나머지 값의 확률은 낮아진다면 엔트로피가 작아진다. 반대로 여러가지 값이 나올 확률이 대부분 비슷한 경우에는 엔트로피가 높아진다. 엔트로피는 확률분포의 모양이 어떤지를 나타내는 특성값 중 하나로 볼 수도 있다. 수학적으로 엔트로피는 확률분포함수를 입력으로 받아 숫자를 출력하는 범함수로 정의한다.  

확률변수 X가 베르누이 분포 또는 카테고리 분포 등의 이산확률분포를 따른다면, 엔트로피를 좀 더 명시적으로 나타낼 수 있다. 베르누이 분포의 엔트로피를 성공확률 theta의 함수로 나타내는 것을 이진 엔트로피 함수라고 부른다.  

이진 엔트로피 함수에 따르면, 각 경우의 엔트로피는 다음과 같다.


```
import numpy as np

# p = 0.4  
print(f"p = 0.4일 경우 엔트로피: {(-0.4 * np.log(0.4)) - 0.6 * np.log(0.6)}")
# p = 0.5
print(f"p = 0.5일 경우 엔트로피: {(-0.5 * np.log(0.5)) - 0.5 * np.log(0.5)}")
# p = 0.6
print(f"p = 0.6일 경우 엔트로피: {(-0.6 * np.log(0.6)) - 0.4 * np.log(0.4)}")
```
